```python
def SingleSaveHistPlot(data_frame, column, ch_sim, outputDir="Output", chosen_hue='Nume_Grup'):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    name_c = "Sim"+ str(ch_sim)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c) 
    #aux = data_frame
    aux = data_frame[data_frame["Simulation_name"]==ch_sim]
    seaborn_elem = sns.histplot(data = aux[['cycle',column,chosen_hue]], 
                                    x = 'cycle', y=column,hue=chosen_hue)
    #seaborn_elem.fig.suptitle("HistPlot - "+str(ch_sim))
    #seaborn_elem.set_title("HistPlot - "+str(ch_sim))
    fig = seaborn_elem.get_figure()
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    #fig.set_title("HistPlot - "+str(ch_sim))
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close()

def SingleSaveHistPlot_bins(data_frame, column, ch_sim, outputDir="Output", chosen_hue='Nume_Grup', _bins=30):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    name_c = "Sim"+ str(ch_sim)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c) 
    #aux = data_frame
    aux = data_frame[data_frame["Simulation_name"]==ch_sim]
    seaborn_elem = sns.histplot(data = aux[['cycle',column,chosen_hue]], 
                                    x = 'cycle', y=column,hue=chosen_hue, bins=_bins)
    #seaborn_elem.fig.suptitle("HistPlot - "+str(cycle))
    #seaborn_elem.set_title("HistPlot - "+str(cycle))
    fig = seaborn_elem.get_figure()
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close()

def SingleSaveJoinPlot(data_frame, column, ch_sim, outputDir="Output", chosen_kind='scatter'):
    name_c = "Sim"+ str(ch_sim)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c) 
    #aux = data_frame
    aux = data_frame[data_frame["Simulation_name"]==ch_sim]
    seaborn_elem = sns.jointplot(x = 'cycle', y=column, 
                                     data = aux[['cycle',column]],kind = chosen_kind)
    seaborn_elem.fig.suptitle("JoinPlot - "+str(ch_sim))
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    
    seaborn_elem.fig.savefig(dir_name, bbox_inches="tight")
    
    plt.clf()
    plt.close()
    
def SingleSaveJoinPlot_bins(data_frame, column, ch_sim, outputDir="Output", chosen_kind='scatter', bins=30):
    name_c = "Sim "+ str(ch_sim)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c) 
    #aux = data_frame
    aux = data_frame[data_frame["Simulation_name"]==ch_sim]
    seaborn_elem = sns.jointplot(x = 'cycle', y=column, 
                                     data = aux[['cycle',column]],kind = chosen_kind, s=bins)
    seaborn_elem.fig.suptitle("JoinPlot - "+str(ch_sim))
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    
    seaborn_elem.fig.savefig(dir_name, bbox_inches="tight")
    
    plt.clf()
    plt.close()
    
def SingleSaveScatterPlot(data_frame, column, ch_sim, outputDir="Output", chosen_hue='Nume_Grup'):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    name_c = "Sim"+ str(ch_sim)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c) 
        
    #aux = data_frame
    aux = data_frame[data_frame["Simulation_name"]==ch_sim]
    seaborn_elem = sns.scatterplot(data = aux[['cycle',column,chosen_hue]], 
                                    x = 'cycle', y=column,hue=chosen_hue)
    seaborn_elem.set_title("ScatterPlot - "+str(name_c))
    fig = seaborn_elem.get_figure()
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close()

def SingleSaveKdePlot(data_frame, column, ch_sim, outputDir="Output"):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    name_c = "Sim"+ str(ch_sim)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c)
    #aux = data_frame
    aux = data_frame[data_frame["Simulation_name"]==ch_sim]
    seaborn_elem = sns.kdeplot(x=aux['cycle'], y=aux[column])
    seaborn_elem.set_title("KdePlot - "+str(ch_sim))
    fig = seaborn_elem.get_figure()
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close()

def SingleSaveFacetGridPlot(data_frame, column, ch_sim, outputDir="Output", chosen_hue='Type'):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    name_c = "Sim"+ str(ch_sim)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c)
    #aux = data_frame
    aux = data_frame[data_frame["Simulation_name"]==ch_sim]
    seaborn_elem = sns.FacetGrid(data=aux[[chosen_hue, column, 'cycle']],col=chosen_hue) 
    seaborn_elem.fig.suptitle("FacetGrid - "+str(ch_sim))
    seaborn_elem.map(sns.scatterplot,'cycle',column)
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    seaborn_elem.fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close() 
```

```python
def SingleSaveHeatMap_v1(data_frame, ch_sim, outputDir="Output"):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    #aux = data_frame
    aux = data_frame[data_frame["Simulation_name"]==ch_sim]
    
    seaborn_elem = sns.heatmap(aux[list(aux.columns)[11:]].corr(), annot=False)
    seaborn_elem.set_title("HeatMap - "+str(ch_sim))
    fig = seaborn_elem.get_figure()
    plt.figure(figsize = (12,12))
    dir_name = os.path.join(os.path.join(os.getcwd(),outputDir),str(ch_sim) + ".png")
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close() 
    
def SingleSaveHeatMap_v1_output(data_frame, ch_sim, outputDir="Output"):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    #aux = data_frame
    aux = data_frame[data_frame["Simulation_name"]==ch_sim]
    
    seaborn_elem = sns.heatmap(aux[list(aux.columns)[11:]].corr(), annot=False)
    seaborn_elem.set_title("HeatMap - "+str(ch_sim))
```

```python

    
def SingleSaveHeatMap_v2_cycle(data_frame, column, ch_sim, outputDir="Output", pt_column='Nume_Grup'):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    aux = data_frame[data_frame["Simulation_name"]==ch_sim]


    pt1 = aux.pivot_table(index='cycle', columns=pt_column,
                                     values=column )
    seaborn_elem = sns.heatmap(pt1,cmap='coolwarm', fmt='.2f')
    seaborn_elem.set_title("HeatMap - "+str(ch_sim) + ' - ' + str(column))
    name_c = "Sim"+str(ch_sim)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c)    
    fig = seaborn_elem.get_figure()
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close()  
```
