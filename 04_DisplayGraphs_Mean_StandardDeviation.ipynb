```python
def SingleSaveHistPlot(data_frame, column, cycle, outputDir="Output", chosen_hue='Type'):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    name_c = "Cycle "+ str(cycle)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c) 
    #aux = data_frame
    aux = data_frame[data_frame["cycle"]==cycle]
    seaborn_elem = sns.histplot(data = aux[['FaultLocation_FromGenerator',column,chosen_hue]], 
                                    x = 'FaultLocation_FromGenerator', y=column,hue=chosen_hue)
    sns.move_legend(seaborn_elem, "upper left", bbox_to_anchor=(1, 1))
    #seaborn_elem.fig.suptitle("HistPlot - Cycle "+str(cycle))
    #seaborn_elem.set_title("HistPlot - Cycle "+str(cycle))
    fig = seaborn_elem.get_figure()
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    #fig.set_title("HistPlot - Cycle "+str(cycle))
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close()

def SingleSaveHistPlot_bins(data_frame, column, cycle, outputDir="Output", chosen_hue='Type', _bins=30):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    name_c = "Cycle "+ str(cycle)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c) 
    #aux = data_frame
    aux = data_frame[data_frame["cycle"]==cycle]
    seaborn_elem = sns.histplot(data = aux[['FaultLocation_FromGenerator',column,chosen_hue]], 
                                    x = 'FaultLocation_FromGenerator', y=column,hue=chosen_hue, bins=_bins)
    sns.move_legend(seaborn_elem, "upper left", bbox_to_anchor=(1, 1))
    #seaborn_elem.fig.suptitle("HistPlot - Cycle "+str(cycle))
    #seaborn_elem.set_title("HistPlot - Cycle "+str(cycle))
    fig = seaborn_elem.get_figure()
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close()

def SingleSaveJoinPlot(data_frame, column, cycle, outputDir="Output", chosen_kind='scatter'):
    name_c = "Cycle "+ str(cycle)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c) 
    #aux = data_frame
    aux = data_frame[data_frame["cycle"]==cycle]
    seaborn_elem = sns.jointplot(x = 'FaultLocation_FromGenerator', y=column, 
                                     data = aux[['FaultLocation_FromGenerator',column]],kind = chosen_kind)
    seaborn_elem.fig.suptitle("JoinPlot - Cycle "+str(cycle))
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    
    seaborn_elem.fig.savefig(dir_name, bbox_inches="tight")
    
    plt.clf()
    plt.close()
    
def SingleSaveJoinPlot_bins(data_frame, column, cycle, outputDir="Output", chosen_kind='scatter', bins=30):
    name_c = "Cycle "+ str(cycle)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c) 
    #aux = data_frame
    aux = data_frame[data_frame["cycle"]==cycle]
    seaborn_elem = sns.jointplot(x = 'FaultLocation_FromGenerator', y=column, 
                                     data = aux[['FaultLocation_FromGenerator',column]],kind = chosen_kind, s=bins)
    seaborn_elem.fig.suptitle("JoinPlot - Cycle "+str(cycle))
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    
    seaborn_elem.fig.savefig(dir_name, bbox_inches="tight")
    
    plt.clf()
    plt.close()
    
def SingleSaveScatterPlot(data_frame, column, cycle, outputDir="Output", chosen_hue='Type'):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    name_c = "Cycle "+ str(cycle)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c) 
        
    #aux = data_frame
    aux = data_frame[data_frame["cycle"]==cycle]
    seaborn_elem = sns.scatterplot(data = aux[['FaultLocation_FromGenerator',column,chosen_hue]], 
                                    x = 'FaultLocation_FromGenerator', y=column,hue=chosen_hue)
    sns.move_legend(seaborn_elem, "upper left", bbox_to_anchor=(1, 1))
    seaborn_elem.set_title("ScatterPlot - Cycle "+str(cycle))
    fig = seaborn_elem.get_figure()
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close()

def SingleSaveKdePlot(data_frame, column, cycle, outputDir="Output"):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    name_c = "Cycle "+ str(cycle)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c)
    #aux = data_frame
    aux = data_frame[data_frame["cycle"]==cycle]
    seaborn_elem = sns.kdeplot(x=aux['FaultLocation_FromGenerator'], y=aux[column])
    seaborn_elem.set_title("KdePlot - Cycle "+str(cycle))
    fig = seaborn_elem.get_figure()
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close()

def SingleSaveFacetGridPlot(data_frame, column, cycle, outputDir="Output", chosen_hue='Type'):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    name_c = "Cycle "+ str(cycle)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c)
    #aux = data_frame
    aux = data_frame[data_frame["cycle"]==cycle]
    seaborn_elem = sns.FacetGrid(data=aux[[chosen_hue, column, 'FaultLocation_FromGenerator']],col=chosen_hue) 
    seaborn_elem.fig.suptitle("FacetGrid - Cycle "+str(cycle))
    seaborn_elem.map(sns.scatterplot,'FaultLocation_FromGenerator',column)
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    seaborn_elem.fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close() 
```

```python
def SingleSaveHeatMap_v1(data_frame, cycle, outputDir="Output"):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    #aux = data_frame
    aux = data_frame[data_frame["cycle"]==cycle]
    
    seaborn_elem = sns.heatmap(aux[list(aux.columns)[11:]].corr(), annot=False)
    seaborn_elem.set_title("HeatMap - Cycle "+str(cycle))
    fig = seaborn_elem.get_figure()
    plt.figure(figsize = (12,12))
    dir_name = os.path.join(os.path.join(os.getcwd(),outputDir),"Cycle"+str(cycle) + ".png")
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close() 
    
def SingleSaveHeatMap_v1_output(data_frame, cycle, outputDir="Output"):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    #aux = data_frame
    aux = data_frame[data_frame["cycle"]==cycle]
    
    seaborn_elem = sns.heatmap(aux[list(aux.columns)[8:]].corr(), annot=False)
    seaborn_elem.set_title("HeatMap - Cycle "+str(cycle))
```

```python
def SingleSaveHeatMap_v2_phase(data_frame, column, outputDir="Output", ch_phase='A', pt_column='cycle'):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    aux = data_frame[data_frame["phase"]==ch_phase]
    copy = aux.copy()
    copy['FaultLocation_FromGenerator'] = aux['FaultLocation_FromGenerator'].apply(lambda x: np.round(x,2))
    
    pt1 = copy.pivot_table(index='FaultLocation_FromGenerator', columns=pt_column,
                                     values=column )
    seaborn_elem = sns.heatmap(pt1,cmap='coolwarm', fmt='.2f')
    seaborn_elem.set_title("HeatMap - Phase "+str(ch_phase) + ' - ' + str(column))
    fig = seaborn_elem.get_figure()
    name_c = "Phase"+ch_phase
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c)
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close()
    
def SingleSaveHeatMap_v2_cycle(data_frame, column, outputDir="Output", ch_cycle = 19, pt_column='phase'):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    aux = data_frame[data_frame["cycle"]==ch_cycle]
    copy = aux.copy()
    copy['FaultLocation_FromGenerator'] = aux['FaultLocation_FromGenerator'].apply(lambda x: np.round(x,2))

    pt1 = copy.pivot_table(index='FaultLocation_FromGenerator', columns=pt_column,
                                     values=column )
    seaborn_elem = sns.heatmap(pt1,cmap='coolwarm', fmt='.2f')
    seaborn_elem.set_title("HeatMap - Cycle "+str(ch_cycle) + ' - ' + str(column))
    name_c = "Cycle"+str(ch_cycle)
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c)    
    fig = seaborn_elem.get_figure()
    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close()  

def SingleSaveHeatMap_v2_type(data_frame, column, outputDir="Output", ch_type = '', pt_column='cycle'):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    aux = data_frame[data_frame["Type"]==ch_type]
    copy = aux.copy()
    copy['FaultLocation_FromGenerator'] = aux['FaultLocation_FromGenerator'].apply(lambda x: np.round(x,2)) 

    pt1 = copy.pivot_table(index='FaultLocation_FromGenerator', columns=pt_column,
                                     values=column )
    seaborn_elem = sns.heatmap(pt1,cmap='coolwarm', fmt='.2f')
    seaborn_elem.set_title("HeatMap - Type "+str(ch_type) + ' - ' + str(column))
    fig = seaborn_elem.get_figure()
    name_c = "Type"+ch_type
    
    isExist = os.path.exists("./"+outputDir+"/"+name_c)
    if not isExist:
        os.makedirs("./"+outputDir+"/"+name_c)

    dir_name = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),name_c),str(column) + ".png")
    fig.savefig(dir_name, bbox_inches="tight")
    plt.clf()
    plt.close()  
```
