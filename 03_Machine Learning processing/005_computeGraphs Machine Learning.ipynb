```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import os

import path
import math
import cmath
import shutil
import pywt
import csv
```

```python
data_pred_1 = pd.read_excel('Precision.xlsx',sheet_name='Error_1')
data_pred_5 = pd.read_excel('Precision.xlsx',sheet_name='Error_5')
data_pred_10 = pd.read_excel('Precision.xlsx',sheet_name='Error_10')
```

```python
allModels = list(data_pred_1["Model"].unique())
allData = list(data_pred_1["data Input"].unique())
allColumns = list(data_pred_1["columns"].unique())
```

```python
data = data_pred_1
dirName = 'data_pred_1'

#data = data_pred_5
#dirName = 'data_pred_5'

#data = data_pred_10
#dirName = 'data_pred_10'
```

```python
for model in allModels:
    for dataInput in allData:
        for _colType in allColumns:
            _aux1 = data_pred_1[(data_pred_1["Model"]==str(model)) 
                        & (data_pred_1["data Input"]==str(dataInput)) 
                        & (data_pred_1["columns"]==str(_colType))]
            _aux5 = data_pred_5[(data["Model"]==str(model)) 
                        & (data_pred_5["data Input"]==str(dataInput)) 
                        & (data_pred_5["columns"]==str(_colType))]
            _aux10 = data_pred_10[(data["Model"]==str(model)) 
                        & (data_pred_10["data Input"]==str(dataInput)) 
                        & (data_pred_10["columns"]==str(_colType))]            
            if len(_aux1) > 0:
                outputDir = os.path.join(os.path.join(os.path.join(os.getcwd(), str(model)), str(dataInput)),str(_colType))
                isExist = os.path.exists(outputDir)
                if not isExist:
                    os.makedirs(outputDir)
                x = list(range(10,20))
                y1 = list()
                y5 = list()
                y10 = list()
                for _x in x:
                    y1.append(_aux1[_x])
                    y5.append(_aux5[_x])
                    y10.append(_aux10[_x])
                fig, axes = plt.subplots(1, 1, figsize=(6,6))
                axes.plot(x, y1)
                axes.set_title('Data input: '+ str(dataInput) + '\n' + 
                                 'Studied columns: '+ str(_colType) + '\n'+ 
                                 'Precision - 1%')
                
                plt.xlabel('#No. Cycles')
                plt.ylabel("Precision - 1%")
                    
                fig.savefig(os.path.join(outputDir, 'Error_1.png'), bbox_inches="tight")
                plt.clf()
                plt.close()                
                fig, axes = plt.subplots(1, 1, figsize=(6,6))
                axes.plot(x, y5)
                axes.set_title('Data input: '+ str(dataInput) + '\n' + 
                                 'Studied columns: '+ str(_colType) + '\n'+ 
                             'Precision - 5%')
                
                plt.xlabel('#No. Cycles')
                plt.ylabel("Precision - 5%")
                    
                fig.savefig(os.path.join(outputDir, 'Error_5.png'), bbox_inches="tight")
                plt.clf()
                plt.close()
                fig, axes = plt.subplots(1, 1, figsize=(6,6))
                axes.plot(x, y10)
                axes.set_title('Data input: '+ str(dataInput) + '\n' + 
                                 'Studied columns: '+ str(_colType) + '\n'+ 
                                'Precision - 10%')
                
                plt.xlabel('#No. Cycles')
                plt.ylabel("Precision - 10%")
                    
                fig.savefig(os.path.join(outputDir, 'Error_10.png'), bbox_inches="tight")
                
                plt.clf()
                plt.close()      
                
                
```

```python
for model in allModels:
    for dataInput in allData:
        _aux = data[(data["Model"]==str(model)) 
                        & (data["data Input"]==str(dataInput)) 
                        ]
        if len(_aux) > 0:
            outputDir = os.path.join(os.path.join(os.path.join(os.getcwd(), str(title)), str(model)),str(dataInput))
            isExist = os.path.exists(outputDir)
            if not isExist:
                    os.makedirs(outputDir)
            for _n in range(10,20):
                seaborn_elem = sns.boxplot(data = _aux[["columns",_n]], 
                                    x = "columns", y=_n )
                plt.xticks(rotation=90)
                seaborn_elem.set_title("ScatterPlot - Cycle "+str(_n))
                
                seaborn_elem.set_title('Model: '+str(model)+'\n'
                              + 'Data input: '+ str(dataInput) + '\n'
                              + str(title) + '\n' 
                              + 'Cycle: ' + str(_n) )
                plt.xlabel('Chosen columns in the model')
                plt.ylabel(str(title))
                fig = seaborn_elem.get_figure()
                fig.savefig(os.path.join(outputDir,'Cycle_' + str(_n)+'.png'), bbox_inches="tight")
                plt.clf()
                plt.close()
```
