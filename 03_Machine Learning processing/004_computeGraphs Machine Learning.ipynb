```python
allModels_coef = list(data_coeff["Model"].unique())
allModels_met = list(data_Metrics["Model"].unique())
all_dataInput = list(data_coeff["data Input"].unique())
all_Columns = list(data_coeff["columns"].unique())
list_metrics = list(data_Metrics["Value"].unique())
dict_metrics={'model.intercept_':'INTERCEPT', 
             'metrics.mean_absolute_error(y_test,predictions)':'MEAN ABSOLUTE ERROR',
             'metrics.mean_squared_error(y_test,predictions)': 'MEAN SQUARE ERROR',
             'np.sqrt(metrics.mean_squared_error(y_test,predictions))':'ROOT MEAN SQUARE ERROR'}
scaled_list=['scaled_data_normal','1_ordin_scaled_data_normal','2_ordin_scaled_data_normal']
normaml_list=['data_normal','1_ordin_data_normal','2_ordin_data_normal']
```

```python
for model in allModels_met:
    for dataInput in all_dataInput:
        for _colType in all_Columns:
            for _met in list_metrics:
                _data = data_Metrics[(data_Metrics["Model"]==str(model)) 
             & (data_Metrics["data Input"]==str(dataInput))
            & (data_Metrics["columns"]==str(_colType))
             & (data_Metrics["Value"]==str(_met))]
                if len(_data) > 0:
                    #outputDir = os.path.join(os.path.join(os.path.join(os.path.join(os.getcwd(), str(model)), str(dataInput)),str(_colType)), str(_met))
                    outputDir = os.path.join(os.path.join(os.path.join(os.getcwd(), str(model)), str(dataInput)),str(_colType))
                    isExist = os.path.exists(outputDir)
                    if not isExist:
                        os.makedirs(outputDir)
                    _title = dict_metrics[_met]
                    x = list(range(10,20))
                    y = list()
                    for _x in x:
                        y.append(_data[_x])
                    fig, axes = plt.subplots(1, 1, figsize=(6,6))
                    axes.plot(x, y)
                    axes.set_title('Data input: '+ str(dataInput) + '\n' + 
                                 'Studied columns: '+ str(_colType) + '\n'+ str(dict_metrics[_met]))

                    plt.xlabel('#No. Cycles')
                    plt.ylabel(str(dict_metrics[_met]))
                    
                    fig.savefig(os.path.join(outputDir, _title+'.png'), bbox_inches="tight")
                    plt.clf()
                    plt.close()
```
