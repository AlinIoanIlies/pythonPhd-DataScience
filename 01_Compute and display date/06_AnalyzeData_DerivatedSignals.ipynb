```python
def SaveLineGraphsCorrelation(data_frame, outputDir="Output"):
    isExist = os.path.exists(outputDir)
    if not isExist:
        os.makedirs(outputDir)    
    dir_path = os.path.join(os.getcwd(),outputDir)
    phases_list = list(data_frame["Phase"].unique())
    for phase in phases_list:
        aux = data_frame[(data_frame["Phase"] == phase)]
        isExist = os.path.exists(os.path.join(os.path.join(os.getcwd(),outputDir)
                                                      ,phase))
        if not isExist:
            os.makedirs(os.path.join(os.path.join(os.getcwd(),outputDir)
                                                      ,phase))
        for column in list(aux.columns)[3:]:
            fileName = 'Case_'+phase+'_'+column + ".png"
            outputImage = os.path.join(os.path.join(os.path.join(os.getcwd(),outputDir),str(phase)), fileName)
            plt.clf()
            plt.figure()
            axes=plt.axes([0,0,1,1])
            axes.plot(aux["Cycle"],aux[str(column)], linewidth=3)
            axes.set_title('Case: '+ str(phase) + '\n'+ str(column) )
            axes.set_xlabel('Number of Cycles')
            axes.set_ylabel('Correlation Factor')
            axes.grid(True,color='#000000')
            plt.savefig(outputImage)
            plt.close()
```
